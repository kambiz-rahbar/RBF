```python
from sklearn.gaussian_process import GaussianProcessRegressor
from numpy import arange, array
from matplotlib.pyplot import scatter, plot
```

```python
x = arange(-1,1.1,0.1)
x = x.reshape(-1,1)

t = array([-.9602,-0.5770,-0.0729,0.3771,0.6405,0.6600,0.4609,0.1336,-0.2013,-0.4344,-0.5000,-.3930,-0.1647,0.0988,0.3072,0.3960,0.3449,0.1816,-0.0312,-0.2189,-0.3201])
t = t.reshape(-1,1)

display(x.shape)
display(t.shape)
```

```python
model=GaussianProcessRegressor()

clf = model.fit(x,t)
y_pred = clf.predict(x)

scatter(x, t);
plot(x,y_pred,'-y');
```
